```python
import pandas as pd
df=pd.read_csv('/content/2015.csv')
df
```

```python
df.drop('Region',axis=1)
```

```python
y=df['Happiness Score']
y
```

```python
z=df['Happiness Rank']
z
```

```python
df
```

```python
X=df.drop(['Happiness Rank','Happiness Score','Region','Country'],axis=1)
X
```

```python
df.isna().any()
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.35,random_state=30)
```

```python
X
```

```python
reg.fit(X_train,y_train)
```

```python
X.shape
```

```python
y_test
```

```python
y_test.shape
```

```python
ypre=reg.predict(X_test)
ypre
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,ypre)
```

```python
Results=pd.DataFrame(columns=['Price','Predicted'])
# Results['Price']=y_test
Results['Predicted']=ypre
Results['Predicted'].sort_values(ascending=False)
```
